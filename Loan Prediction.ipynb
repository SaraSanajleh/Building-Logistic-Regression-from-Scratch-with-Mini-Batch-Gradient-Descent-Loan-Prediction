```python
# Normalization for make all data at the same scale
def mm_normalize(df, mn_val, mx_val):

    mn_mat = np.full(df.shape, mn_val)
    mx_mat = np.full(df.shape, mx_val)
    return (df - mn_mat) / (mx_mat - mn_mat)

mn_val = tr.min()
mx_val = tr.max()

tr = mm_normalize(tr, mn_val, mx_val)
tst = mm_normalize(tst, mn_val, mx_val)
val = mm_normalize(val, mn_val, mx_val)
```

# Logistic Regression Model 

```python
''' This code performs binary classification using a logistic regression model with mini batch training.
It includes steps for training, evaluation, and prediction. 
The model is applied to three datasets: tr for training, val for validation, and tst for testing. '''

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def accuracy(y_true, y_pred):  
    y_pred_class = [1 if p >= 0.5 else 0 for p in y_pred]
    return sum(y_true[i] == y_pred_class[i] for i in range(len(y_true))) / len(y_true)

def fit(X_train, y_train, X_val, y_val, b, lr, ep, flag=True):
    bias = 0  
    w = np.random.rand(X_train.shape[1])  
    tr_loss = [] 
    val_loss = []  
    tr_accuracy = []  
    val_accuracy = []  
    bestval_accuracy = 0  
    best_w = w.copy()  
    best_b = bias  

    for epoch in range(ep):
        batch_loss_sum = 0
        num_batches = int(np.ceil(X_train.shape[0] / b))

        for batch in range(num_batches):
            start_idx = batch * b
            end_idx = start_idx + b
            idx = slice(start_idx, end_idx)

            linear_output = np.dot(X_train[idx], w) + bias
            y_hat = sigmoid(linear_output)

            batch_loss = -np.mean(y_train[idx] * np.log(y_hat + 1e-15) + (1 - y_train[idx]) * np.log(1 - y_hat + 1e-15))
            batch_loss_sum += batch_loss

            bias_derivative = -np.mean(y_train[idx] - y_hat)
            bias -= lr * bias_derivative

            w_derivative = -np.dot((y_train[idx] - y_hat), X_train[idx]) / b
            w -= lr * w_derivative

        if epoch % 100 == 0 and epoch > 0:
            lr *= 0.9
            
        tr_loss_avg = batch_loss_sum / num_batches
        tr_loss.append(tr_loss_avg)

        val_predictions = sigmoid(np.dot(X_val, w) + bias)
        val_loss_val = -np.mean(y_val * np.log(val_predictions + 1e-15) + (1 - y_val) * np.log(1 - val_predictions + 1e-15))
        val_loss.append(val_loss_val)

        tr_predictions = sigmoid(np.dot(X_train, w) + bias)
        tr_accuracy_val = accuracy(y_train, tr_predictions)
        tr_accuracy.append(tr_accuracy_val)

        val_accuracy_val = accuracy(y_val, val_predictions)
        val_accuracy.append(val_accuracy_val)

        if val_accuracy_val > bestval_accuracy:
            bestval_accuracy = val_accuracy_val
            best_w = w.copy()
            best_b = bias

        if flag:
            print(f'Epoch {epoch + 1}, Training Loss: {tr_loss_avg:.2f}, Validation Loss: {val_loss_val:.2f}, Training Accuracy: {tr_accuracy_val:.2f}, Validation Accuracy: {val_accuracy_val:.2f}')

    print(f"Best Validation Accuracy: {bestval_accuracy:.2f}")

    return tr_loss, val_loss, tr_accuracy, val_accuracy, best_w, best_b


def predict(X, w, bias, threshold=0.5):
    linear_output = np.dot(X, w) + bias
    probabilities = sigmoid(linear_output)
    return np.array([1 if p >= threshold else 0 for p in probabilities])

X_train = tr.drop(columns=['Loan_Status']).values
y_train = tr['Loan_Status'].values
X_val = val.drop(columns=['Loan_Status']).values
y_val = val['Loan_Status'].values
X_test = tst.drop(columns=['Loan_Status']).values
y_test = tst['Loan_Status'].values

b = 16
lr = 0.01
ep = 589

tr_loss, val_loss, tr_accuracy, val_accuracy, final_w, final_bias = fit(X_train, y_train, X_val, y_val, b, lr, ep)

y_pred_train = predict(X_train, final_w, final_bias)
tr_accuracy_val = accuracy(y_train, y_pred_train)
print(f"Training Accuracy: {tr_accuracy_val:.2f}")

y_pred_val = predict(X_val, final_w, final_bias)
val_accuracy_val = accuracy(y_val, y_pred_val)
print(f"Validation Accuracy: {val_accuracy_val:.2f}")

y_pred_test = predict(X_test, final_w, final_bias)
test_accuracy = accuracy(y_test, y_pred_test)
print(f"Test Accuracy: {test_accuracy:.2f}")

print("Training Predictions:", y_pred_train)
print("Validation Predictions:", y_pred_val)
print("Test Predictions:", y_pred_test)
```

```python
# this code tests a number of different training epochs to determine the most appropriate one based on the model's validation accuracy.
def experiment_epochs(X_train, y_train, X_val, y_val, batch_size, learning_rate, epoch_values):
    best_val_accuracy = 0
    best_max_epochs = 0

    for max_epochs in epoch_values:
        print(f"\nTraining with {max_epochs} epochs...")

        train_loss, val_loss, train_accuracy_history, val_accuracy_history, final_coef, final_intercept = fit(
            X_train, y_train, X_val, y_val, batch_size, learning_rate, max_epochs, flag=False
        )

        val_predictions = predict(X_val, final_coef, final_intercept)
        val_accuracy =accuracy(y_val, val_predictions)

        print(f"Validation Accuracy for {max_epochs} epochs: {val_accuracy:.2f}")

        if val_accuracy > best_val_accuracy:
            best_val_accuracy = val_accuracy
            best_max_epochs = max_epochs

    print(f"\nBest result obtained with {best_max_epochs} epochs: Validation Accuracy = {best_val_accuracy:.2f}")

epoch_values = [670,789,589,538,629]

experiment_epochs(X_train, y_train, X_val, y_val, batch_size=16, learning_rate=0.01, epoch_values=epoch_values)
```

```python
# This is the same actual model but hust i add regularization
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def accuracy(y_true, y_pred):  
    y_pred_class = [1 if p >= 0.5 else 0 for p in y_pred]
    return sum(y_true[i] == y_pred_class[i] for i in range(len(y_true))) / len(y_true)

def fit(X_train, y_train, X_val, y_val, b, lr, ep, lambda_, flag=True):
    bias = 0  
    w = np.random.rand(X_train.shape[1])  
    tr_loss = [] 
    val_loss = []  
    tr_accuracy = []  
    val_accuracy = []  
    bestval_accuracy = 0  
    best_w = w.copy()  
    best_b = bias  

    for epoch in range(ep):
        batch_loss_sum = 0
        num_batches = int(np.ceil(X_train.shape[0] / b))

        for batch in range(num_batches):
            start_idx = batch * b
            end_idx = start_idx + b
            idx = slice(start_idx, end_idx)

            linear_output = np.dot(X_train[idx], w) + bias
            y_hat = sigmoid(linear_output)

            batch_loss = -np.mean(y_train[idx] * np.log(y_hat + 1e-15) + (1 - y_train[idx]) * np.log(1 - y_hat + 1e-15)) + (lambda_ / (2 * X_train.shape[0])) * np.sum(w ** 2)
            batch_loss_sum += batch_loss

            bias_derivative = -np.mean(y_train[idx] - y_hat)
            bias -= lr * bias_derivative

            w_derivative = -np.dot((y_train[idx] - y_hat), X_train[idx]) / b + (lambda_ * w) / b
            w -= lr * w_derivative

        if epoch % 100 == 0 and epoch > 0:
            lr *= 0.9
        tr_loss_avg = batch_loss_sum / num_batches
        tr_loss.append(tr_loss_avg)

        val_predictions = sigmoid(np.dot(X_val, w) + bias)
        val_loss_val = -np.mean(y_val * np.log(val_predictions + 1e-15) + (1 - y_val) * np.log(1 - val_predictions + 1e-15)) + (lambda_ / (2 * X_val.shape[0])) * np.sum(w ** 2)
        val_loss.append(val_loss_val)

        tr_predictions = sigmoid(np.dot(X_train, w) + bias)
        tr_accuracy_val = accuracy(y_train, tr_predictions)
        tr_accuracy.append(tr_accuracy_val)

        val_accuracy_val = accuracy(y_val, val_predictions)
        val_accuracy.append(val_accuracy_val)

        if val_accuracy_val > bestval_accuracy:
            bestval_accuracy = val_accuracy_val
            best_w = w.copy()
            best_b = bias

        if flag:
            print(f'Epoch {epoch + 1}, Training Loss: {tr_loss_avg:.2f}, Validation Loss: {val_loss_val:.2f}, Training Accuracy: {tr_accuracy_val:.2f}, Validation Accuracy: {val_accuracy_val:.2f}')

    print(f"Best Validation Accuracy: {bestval_accuracy:.2f}")

    return tr_loss, val_loss, tr_accuracy, val_accuracy, best_w, best_b


def predict(X, w, bias, threshold=0.5):
    linear_output = np.dot(X, w) + bias
    probabilities = sigmoid(linear_output)
    return np.array([1 if p >= threshold else 0 for p in probabilities])

X_train = tr.drop(columns=['Loan_Status']).values
y_train = tr['Loan_Status'].values
X_val = val.drop(columns=['Loan_Status']).values
y_val = val['Loan_Status'].values
X_test = tst.drop(columns=['Loan_Status']).values
y_test = tst['Loan_Status'].values

b = 16
lr = 0.01
ep = 589
lambda_ = 0.001 

tr_loss, val_loss, tr_accuracy, val_accuracy, final_w, final_bias = fit(X_train, y_train, X_val, y_val, b, lr, ep, lambda_)

y_pred_train = predict(X_train, final_w, final_bias)
tr_accuracy_val = accuracy(y_train, y_pred_train)
print(f"Training Accuracy: {tr_accuracy_val:.2f}")

y_pred_val = predict(X_val, final_w, final_bias)
val_accuracy_val = accuracy(y_val, y_pred_val)
print(f"Validation Accuracy: {val_accuracy_val:.2f}")

y_pred_test = predict(X_test, final_w, final_bias)
test_accuracy = accuracy(y_test, y_pred_test)
print(f"Test Accuracy: {test_accuracy:.2f}")

print("Training Predictions:", y_pred_train)
print("Validation Predictions:", y_pred_val)
print("Test Predictions:", y_pred_test)
```
